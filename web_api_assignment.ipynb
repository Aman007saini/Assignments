# Web API Assignment

## Q1. What is an API? Give an example, where an API is used in real life.

**API (Application Programming Interface)** is a set of rules and protocols that allows different software applications to communicate with each other.  
**Example:** When you use a weather app on your phone, it uses an API to fetch weather data from a remote server.

---

## Q2. Give advantages and disadvantages of using API.

**Advantages:**
- Enables communication between different software systems.
- Promotes code reusability and modularity.
- Simplifies integration with third-party services.
- Enhances security by exposing only necessary data/functions.

**Disadvantages:**
- Can introduce security vulnerabilities if not properly managed.
- Versioning and backward compatibility can be challenging.
- Dependency on external APIs may lead to issues if the API changes or is unavailable.

---

## Q3. What is a Web API? Differentiate between API and Web API.

**Web API** is an API that can be accessed over the web using HTTP/HTTPS protocols.

**Difference:**
- **API:** General term for any interface that allows communication between software components.
- **Web API:** A specific type of API accessible via web protocols (HTTP/HTTPS).

---

## Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

**REST (Representational State Transfer):**
- Architectural style for designing networked applications.
- Uses standard HTTP methods (GET, POST, PUT, DELETE).
- Stateless and supports multiple data formats (JSON, XML).

**SOAP (Simple Object Access Protocol):**
- Protocol for exchanging structured information in web services.
- Uses XML for message format.
- Requires strict standards and is more complex.

**Shortcomings of SOAP:**
- More complex and heavyweight compared to REST.
- Slower due to XML parsing.
- Less flexible with data formats.
